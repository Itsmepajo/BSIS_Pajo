```python

from transformers import pipeline


qa = pipeline("question-answering")

context = """
The Amazon rainforest is the largest tropical rainforest in the world,
covering over five and a half million square kilometers.
It is known for its biodiversity and is home to many unique species.
"""

question = "What is the largest tropical rainforest in the world?"


answer = qa(question=question, context=context)

print(f"Question: {question}")
print(f"Answer: {answer['answer']} (Score: {answer['score']:.3f})")
```

```python
from transformers import pipeline

nlp=pipeline("sentiment-analysis")
texts=[
    "I love using NLP models!",
    "This movie was terrible and boring.",
    "The food was okay, not the best but not bad either."
]

for text in texts:
  result=nlp(text)[0]
  print(f"Text: {text}")
  print(f"Sentiment: {result ['label']}, Score: {result['score']:.4f}")
  print("-" * 40)
```
